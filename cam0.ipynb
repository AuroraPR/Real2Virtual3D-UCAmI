```python
import cv2
import time

import numpy as np
import os
import glob

path ="./out/"
path_fuzzy =path+"cam0/"



#files = glob.glob(path_fuzzy+"*")
#for f in files:
#    os.remove(f)

#files = glob.glob(path_raw+"*")
#for f in files:
#    os.remove(f)


files = glob.glob(path_fuzzy+"*")
for f in files:
    os.remove(f)    

print("GO")
capture = cv2.VideoCapture(0)
print("DONE,",capture.isOpened())

#count=0
#count=0
lastts = int(time.time())
    
while (capture.isOpened()):
    
    ret, frame = capture.read()
    
    write=False
    ts=int(time.time())
    if(ts != lastts):
        lastts=ts
        print("Time new :",lastts)
        write=True
        
   
#    cv2.imwrite(path_fuzzy+str(ts)+".jpg", frame, [int(cv2.IMWRITE_JPEG_QUALITY), 10])
    if(write):
        print('Restaured Dimensions : ',frame.shape)
        cv2.imwrite(path_fuzzy+str(int(ts))+".png", frame, [int(cv2.IMWRITE_PNG_COMPRESSION), 9])
    

    cv2.imshow('webCam',frame)
    
    keyCode = cv2.waitKey(1)
    if (keyCode == ord('s')):
            break


    if cv2.getWindowProperty('webCam', cv2.WND_PROP_VISIBLE) <1:
        break
        
    #count=count+1
    #if(count > 100):
    #    break
        
    #time.sleep(3)        

capture.release()
cv2.destroyAllWindows()
```
