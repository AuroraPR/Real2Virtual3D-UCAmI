```python
import cv2
import os
import numpy as np
from PIL import Image
  
# importing library for plotting
from matplotlib import pyplot as plt

scene="test/scene5"

path0="./out/"+scene+"/raw_cam2/"
dir_path0=path0+'img/'
dir_path00=path0+'img3D/'
dir_path01=path0+'foot2D/'
dir_path000=path0+'faces/'
dir_path0000=path0+'land3D/'
times=[]
for path in os.listdir(dir_path0):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path0, path)) and not '_' in path:
        times.append(path.replace('.jpg',''))
times.sort()
print(times)
```

```python
from pandas import read_csv   
series=read_csv(path0+"tracking.txt", sep="\t",parse_dates=False)
tracking=series.values


print(tracking[:,1])

def getTracking(id0):
    ix=np.where(tracking[:,1] == id0)[0]
    if ix.size>0:
        return tracking[np.where(tracking[:,1] == id0)][0][2]
    return None

print(getTracking("1689846772_5"))
```

```python
from pandas import read_csv   
series=read_csv(dir_path01+"tracking.txt", sep="\t",parse_dates=False)
foots=series.values


print(foots[:,1])

def getFoot(id0):
    ix=np.where(foots[:,0] == id0)[0]
    if ix.size>0:
        return foots[np.where(foots[:,0] == id0)][0][1:]
    return np.zeros(2)

print(getFoot("1689846s772_5"))
```

```python
from pandas import read_csv   
series=read_csv(dir_path000+"tracking.txt", sep="\t",parse_dates=False)
faces=series.values


print(faces[:,1])

def getFace(id0):
    ix=np.where(faces[:,0] == id0)[0]
    if ix.size>0:
        return faces[np.where(faces[:,0] == id0)][0][1]
    return None

print(getFace("1689846772_5"))
```

```python
from pandas import read_csv   



def get3D(id0):
    if os.path.isfile(os.path.join(dir_path0000, id0+".json")):
        series=read_csv(os.path.join(dir_path0000, id0+".json"), sep="\t",parse_dates=True,header=None)
        data=np.array(series.values)
        #print(series.values)
        
        head=np.mean(data[0:10], axis=0)
        #print("head:",head)
        
        hand_left=np.mean(data[15:22:2], axis=0)
        #print("hand_left:",hand_left)
        hand_right=np.mean(data[16:23:2], axis=0)
        #print("hand_right:",hand_right)
        
        foot_left=np.mean(data[27:32:2], axis=0)
        #print("foot_left:",foot_left)
        foot_right=np.mean(data[28:33:2], axis=0)
        #print("foot_right:",foot_right)

        
        ret=[head,data[11],data[12],data[13],data[14],hand_right,hand_left,data[23],data[24],data[25],data[26],foot_right,foot_left]
        return np.array(ret)
    else:
        return np.zeros((13,4)) 

print((get3D("1689846772_5")))
```

```python
import glob


ffinal=open(path0+"fusion.tsv","w")
for t in times:
    jpgFilenamesList = glob.glob(dir_path00+t+"*.jpg")
    print(t)
    for f in jpgFilenamesList:
        id0=t+"_"+f[f.rfind("_")+len("_"):f.rfind(".")]
        body=(get3D(id0))
        loc2d=getFoot(id0)
        print("\t"+id0+"\t"+str(getTracking(id0))+"\t"+str(getFace(id0))+"\t"+str(body)+"\t"+str(loc2d[0])+"\t"+str(loc2d[1]))
        ffinal.write(str(t)+"\t"+str(id0)+"\t"+str(getTracking(id0))+"\t"+str(getFace(id0)))
        ffinal.write("\t"+str(loc2d[0]).replace('.',',')+"\t"+str(loc2d[1]).replace('.',','))
        for p in body.flatten():
                print(p)
                ffinal.write("\t"+str(p).replace('.',','))
        ffinal.write("\n")
ffinal.close()
```
