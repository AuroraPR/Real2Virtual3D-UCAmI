```python
import cv2
import os
import numpy as np
from PIL import Image
  
# importing library for plotting
from matplotlib import pyplot as plt

scene="scene2"

path0="./out/final_scenes/"+scene+"/raw_cam1/"
dir_path=path0+'img3D/'
dir_path0=path0+'img/'

alphaH=200


times=[]
for path in os.listdir(dir_path0):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path0, path)) and not '_' in path:
        times.append(path.replace('.jpg',''))
times.sort()
print(times)
```

```python
colors = ('b','g','r')
  
gray_zone=range(185,205)

def getH(img):
    

    # find frequency of pixels in range 0-255
    # compute and plot the image histograms
    hrgb=[]
    for i,color in enumerate(colors):
        hist = cv2.calcHist([img],[i],None,[256],[0,256])
        hrgb.append(hist)
        for j in gray_zone:
            hist[j]=[0]
        #plt.plot(hist,color = color)
        #plt.show()
    hrgb=np.array(hrgb)
    return hrgb
```

```python
from sklearn.metrics import mean_squared_error
import math

def sim(A,B):   
    A0=A.flatten()
    B0=B.flatten()
    
    return math.sqrt(mean_squared_error(A0, B0))

def closerH(usersH, h):
    iV=[]
    for ix,h2 in enumerate(usersH):
        iV.append(sim(h2,h))

    iV=np.array(iV)
    return iV
```

```python
import glob


file = open("tracking.txt", "w")


usersH=[]

for time in times:
    
    jpgFilenamesList = glob.glob(dir_path+time+"*.jpg")

    print(time,str(jpgFilenamesList))
    
    if not jpgFilenamesList:
        continue
    hs=[]
    hv=[]
    img_src=[]
    for f in jpgFilenamesList:
        img = cv2.imread(f)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_src.append(img)
        #plt.imshow(img)
        ih,iw,_=img.shape
        print("img.shape:",str(img.shape),float(ih)/float(iw))
        if(float(ih)/float(iw)<1.75):
            print("bad image size:",float(ih)/float(iw))
            continue
        h=getH(img)
        hs.append(closerH(usersH,h))
        hv.append(h)
        print(hs)
    hs=np.array(hs)
    print(hs)
    file.write(time)
    while True:
        if(not hs.size):
            i=len(usersH)-1
            if len(hv)>0:
                usersH.append(hv[0])
                print("New user!")
            break
        else:
            ij_min = np.unravel_index(hs.argmin(), hs.shape)
            i=ij_min[0]
            print("user:",ij_min, hs[ij_min])
            plt.imshow(img_src[i])
            plt.show()
            if (hs[np.unravel_index(hs.argmin(), hs.shape)]>alphaH):
                usersH.append(hv[i])
            usersH[i]=usersH[i]*0.25+h*0.75
            file.write("\t"+str(ij_min))
        hs[i]=100000
        hs[:,i]=100000
        print("hs-",hs)
        if(hs.min()==100000):
            break
        
    file.write("\n")

        
        
file.close()    
```
