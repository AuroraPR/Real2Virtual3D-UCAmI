```python
import numpy as np
from pandas import read_csv    

import os
import glob
import time

from datetime import datetime


def XYZW(array):
    print(array)
    return str(array[0])+"\t"+str(array[1])+"\t"+str(array[2])+"\t"+str(array[3])
dir_path="./out/raw_cam2/land3D/"
out_path="./out/raw_cam2/unity-view/users/auro/"



files = glob.glob(out_path+"*")
for f in files:
    os.remove(f)
        
for path in os.listdir(dir_path):
    if os.path.isfile(os.path.join(dir_path, path)):
        series=read_csv(os.path.join(dir_path, path), sep="\t",parse_dates=True,header=None)
        data=np.array(series.values)
        print(series.values)
        
        head=np.mean(data[0:10], axis=0)
        print("head:",head)
        
        hand_left=np.mean(data[15:22:2], axis=0)
        print("hand_left:",hand_left)
        hand_right=np.mean(data[16:23:2], axis=0)
        print("hand_right:",hand_right)
        
        foot_left=np.mean(data[27:32:2], axis=0)
        print("foot_left:",foot_left)
        foot_right=np.mean(data[28:33:2], axis=0)
        print("foot_right:",foot_right)

        

        f = open(os.path.join(out_path, path), "w")
        f.write(XYZW(head)+"\n")
        f.write(XYZW(data[11])+"\n")
        f.write(XYZW(data[12])+"\n")
        f.write(XYZW(data[13])+"\n")
        f.write(XYZW(data[14])+"\n")
        f.write(XYZW(hand_right)+"\n")
        f.write(XYZW(hand_left)+"\n")
        f.write(XYZW(data[23])+"\n")
        f.write(XYZW(data[24])+"\n")
        f.write(XYZW(data[25])+"\n")
        f.write(XYZW(data[26])+"\n")        
        f.write(XYZW(foot_right)+"\n")
        f.write(XYZW(foot_left)+"\n")
        f.close()
```
